Question #1: Use NumPy to determine the dot product of the following matrices:

$$\begin{bmatrix} 2 & 2 \\ 0 & 3 \\ 0 & 4 \end{bmatrix} \cdot \begin{bmatrix} 2 & 1 & 2 \\ 3 & 2 & 4 \end{bmatrix}$$

```python
import numpy as np

a = np.array([[2,2], [0,3], [0,4]])
b = np.array([[2,1,2], [3,2,4]])
print(*a.dot(b), sep ='\n')
```

Question #2: Prepare a Python program that will determine the dot product, cross product, outer product,
and determinant of two, user-defined 2 × 2 matrices. Your program should include explicit
printing and should utilize the built-in NumPy functions for completing the four linear algebra
operations.

```python
import numpy as np

def error_1():
    print("Invalid input")

list_1 = []
list_2 = []
def innerfinder(n):
    run = 1
    while run < 5:
        while (1 == 1):
            sing_1 = input("Enter your " + str(run) + " value " + "for matrix " + str(n) + " : ")
            try:
                float(sing_1)
                run = run + 1
                break
            except ValueError:
                error_1()
        sing_1 = float(sing_1)
        if n == 1:
            list_1.append(sing_1)
        else:
            list_2.append(sing_1)
            
innerfinder(1)
innerfinder(2)
slip_1 = np.array(list_1)
slip_2 = np.array(list_2)
s1 = slip_1.reshape(2,2)
s2 = slip_2.reshape(2,2)

while 1 == 1:
    print("\nWhich operation would you like to perform? (D for dot, C for cross, O for outer, Det1 for determinate of Matrix 1, or Det2 for determinate of Matrix 2)")
    op = input("")
    if op == "D":
        print(s1.dot(s2))
    elif op == "C":
        print(np.cross(s1,s2))
    elif op == "O":
        print(np.outer(s1,s2))
    elif op == "Det1":
        print(np.linalg.det(s1))
    elif op == "Det2":
        print(np.linalg.det(s2))
    else:
        error_1()
        continue
    break
```

Question #3: Utilize the built-in NumPy functions to print the minimum, maximum, mean, and median
of a user-defined array in Python. Your code should include user input for the dimensions of
the array and should give the user an option for specifying the spacing of the values within
the array. Consider the linspace function for array creation and utilize explicit printing for
the results.

```python
row = int(input("Enter number of rows for array: "))
column = int(input("Enter number of columns for array: "))
maximum = int(input("What is the maximum value you would like in this matrix? "))
minimum = int(input("What is the minimum value you would like in this matrix? "))

space = row * column
mix = np.linspace(minimum, maximum, num = space)
print(mix.reshape(row, column))
```
