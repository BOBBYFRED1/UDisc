```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def odeEuler(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0, len(t) - 1):
        y[n + 1] = y[n] + f(y[n], t[n]) * (t[n + 1] - t[n])
    return y

def finder(f, y0, x1, x2, y1, y2, name):
    t = np.linspace(x1, x2, 21)
    y = odeEuler(f, y0, t)
    yy = odeint(f, y0, t)
    plt.plot(t, y, 'r.-', t, yy, 'b.-')
    plt.legend(['Euler'])
    plt.grid(True)
    plt.axis([x1, x2, y1, y2])
    plt.title(name)
    plt.show()
```

$$\frac{dy(t)}{dt} = −y(t) + 1$$
